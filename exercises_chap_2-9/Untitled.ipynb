```python
import os
import tarfile
import urllib.request

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
```

```python
fetch_housing_data()
```

```python
import pandas as pd

def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv")
    return pd.read_csv(csv_path)
```

```python
housing = load_housing_data()
```

```python
housing.head()
```

```python
housing["income_cat"] = pd.cut(housing["median_income"],
                               bins=[0.0, 1.5, 3.0, 4.5, 6.0, np.inf], 
                               labels=[1, 2, 3, 4, 5])
```

```python
housing.head()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

sss = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_idx, test_idx in sss.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_idx]
    strat_test_set = housing.loc[test_idx]
```

```python
strat_train_set.drop("income_cat", axis=1, inplace=True)
strat_test_set.drop("income_cat", axis=1, inplace=True)
```

```python
strat_train_set.head()
```

```python
strat_test_set.head()
```

```python
strat_train_set["rooms_per_house"] = strat_train_set["total_rooms"] / strat_train_set["households"]
strat_train_set["bedrooms_per_house"] = strat_train_set["total_bedrooms"] / strat_train_set["households"]

strat_test_set["rooms_per_house"] = strat_test_set["total_rooms"] / strat_test_set["households"]
strat_test_set["bedrooms_per_house"] = strat_test_set["total_bedrooms"] / strat_test_set["households"]
```

```python
train_lab = strat_train_set[["median_house_value"]]
strat_train_set.drop(columns=["median_house_value"], inplace=True)
```

```python
test_lab = strat_test_set[["median_house_value"]]
strat_test_set.drop(columns=["median_house_value"], inplace=True)
```

```python
num_attr = ['longitude',
 'latitude',
 'housing_median_age',
 'total_rooms',
 'total_bedrooms',
 'population',
 'households',
 'median_income',
 'median_house_value',
 'rooms_per_house',
 'bedrooms_per_house']
```

```python
cat_attr = ['ocean_proximity']
```

```python
# Let's create a ColumnTransformer with 2 pipelines (1 for numerical and 1 for categorical attributes)
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy="median")), # fill in the missing values
    ('std_scaler', StandardScaler()), # scale the data
])
```

```python
cat_pipe = Pipeline([
    ('hot_enc', OneHotEncoder(sparse=False)),
])
```

```python
col_transf = ColumnTransformer([
    ("num", num_pipe, num_attr),
    ("cat", cat_pipe, cat_attr),
])
```

```python
col_transf.fit(train)
X_train = col_transf.transform(train)
```

```python
X_test = col_transf.transform(test)
```

```python
# X_train - training set
y_train = train_lab.to_numpy()

# X_test - testing set
y_test = test_lab.to_numpy()
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

y_pred = lin_reg.predict(X_train)
lin_mse = mean_squared_error(y_train, y_pred)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
y_pred = lin_reg.predict(X_test)
lin_mse = mean_squared_error(y_test, y_pred)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```
